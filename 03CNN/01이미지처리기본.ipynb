## openCV
이미지를 읽을 때는 `imread` 메서드를 사용  
이미지 데이터를 NumPy 배열로 저장하지만 색 채널의 순서가 R-G-B 순서가 아니라, `B-G-R 순서`이다.  
`resize()` 명령으로 이미지 크기 변환 기능을 제공한다. 그 외에 다양한 기능을 제공한다.   


https://opencv.org/

```python
import cv2
cv2.__version__
```

```python
img = cv2.imread("flower.jpg")
plt.imshow(img)
print(img.shape)
```

```python
# 각 채널을 분리
b, g, r = cv2.split(img)

# b, r을 서로 바꿔서 Merge
img = cv2.merge([r, g, b])

plt.imshow(img)
```

# Tensorflow에서 이미지 처리

```python
import os 
%tensorflow_version 2.x

import tensorflow as tf

os.listdir()

```

```python
import glob
path = glob.glob('./*.jpg')
path
```

```python
path[0]

gfile = tf.io.read_file(path[0])

image = tf.io.decode_image(gfile)
image.shape 

import matplotlib.pyplot as plt

plt.imshow(image)
```
